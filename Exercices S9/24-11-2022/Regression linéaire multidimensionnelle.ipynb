```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from seaborn import heatmap
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Charger le dataset 'movie_metadata.csv'

df = pd.read_csv('movie_metadata.csv')
```

```python
# Regarder s'il ya des valeurs nulles dans le dataset
# Si oui, remplacer les valeurs nulles par des 0

df = df.fillna(0)
```

```python
# Tracer la "heatmap" des correlations entre les colonnes

newdf = df.select_dtypes(include=np.number)
corr = newdf.corr()
plt.figure(figsize=(18,14))
heatmap(corr, annot=True)
plt.show()
```

```python
# Identifier les colonnes qui sont soit des int64 et float64

tdf = df.select_dtypes(include=['int64', 'float64'])
nb_header = tdf.columns.values.tolist()
nb_header
```

```python
# Diviser le dataset en deux datasets plus petits (un de 70%, qu'on appelera "train" et un de 30% qu'on apelera "test" du dataset initial)
# avec la methode train_test_split

train, test = train_test_split(newdf, test_size=0.3, train_size=0.7)
```

```python
# Choisir la colonne "imdb_score" comme target et toutes les autres comme features, et donc définir y_train, X_train, y_test et X_test

y_train = train['imdb_score']
X_train = train.drop('imdb_score', axis=1)
y_test = test['imdb_score']
X_test = test.drop('imdb_score', axis=1)
```

```python
# Utiliser le model linear_model.LinearRegression() de sklearn pour faire la régression linéaire 
# Donner X_train et y_train au model pour qu'il s'entraine

reg = LinearRegression().fit(X_train, y_train)
```

```python
# Regarder le score du modèle pour X_train et y_train

reg.score(X_train, y_train)
```

```python
# Regarder le score du modèle pour X_test et y_trest

reg.score(X_test, y_test)
```
