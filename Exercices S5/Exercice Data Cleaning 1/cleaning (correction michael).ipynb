```python
import pandas as pd
import numpy as np


df = pd.read_csv("cleaning.csv")

to_drop = ["Unnamed: 0"]
df.drop(to_drop, inplace=True, axis=1)
df.describe()
```

```python
df.count() / df.shape[0] * 100
```

```python
df.drop_duplicates(subset=None, keep="first", inplace=False)
```

```python
df.drop("Colors Reported", inplace=True, axis=1)
df.dropna(inplace=True)
df
```

```python
df["Test_state"] = df["State"].apply((lambda x: x[0] + "." + x[1] + "."))
df.head(10)
```

```python
df["Test_City"] = df["City"].apply(lambda x: ".".join([i[:3] for i in x.split(" ") if i]))
df
```

```python
df["Time"] = pd.to_datetime(df["Time"])
df
```

```python
df["test_Time"] = df["Time"].dt.strftime('%d-%m-%Y')
df
```

```python
df = df[["City", "Test_City", "State", "Test_state", "Time", "test_Time", "Shape Reported"]]
df
```
