```python
import pandas as pd

df = pd.read_csv('cleaning.csv', index_col=0)
```

```python
# Printing the number of values per column
header = df.head()
for head in header: print(f'{head}: {df.shape[0] - df[head].isna().sum() - 1}')
```

```python
# Filling empty rows in the 'Shape Reported' column
df['Shape Reported'] = df['Shape Reported'].fillna('VARIOUS')
df
```

```python
# Removing duplicate rows
df2 = df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
print(f'{len(df) - len(df2)} duplicates have been removed')
```

```python
# Renaming colomns
df2.rename(columns={'Shape Reported': 'Forme', 'Colors Reported': 'Couleur'}, inplace=True)
df2
```

```python
# Reindexing columns
df2.reindex(columns=['City', 'Forme', 'Couleur', 'State', 'Time'])
```

```python
DATES, HEURES = [], []

def fonction(str):
    L = str.split(' ')
    date = L[0].split('/')
    new_date = '/'.join([date[1], date[0], date[2]])
    DATES.append(new_date)
    HEURES.append(L[1])
```

```python
df2['Time'].apply(fonction)
```

```python
df2.insert(5, 'Date', value=DATES)
df2.insert(6, 'Heure', value=HEURES)
df2.drop(columns=['Time'])
```

```python
df2.to_csv('cl.csv', sep=';', index=False)
```
